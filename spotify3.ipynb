```python
def track_name_by_energy (df, energy):
    """Devuelve una lista de títulos de canciones que pertenecen a una energía específica."""
    filtered_df = df[df['energy'] == energy]
    return filtered_df['track_name'].tolist()
print(track_name_by_energy(df, 0.5))


```

```python
# haz una grafica de barras que muestre el tiempo total de reproduccion (in milliseconds) por cada artista (artist_name) 
graph = df.groupby('artists')['duration_ms'].sum().sort_values(ascending=False).head(10)
graph.plot(kind='bar', title='Tiempo total de reproducción por artista (Top 10)', ylabel='Duración total (ms)', xlabel='Artista')
import matplotlib.pyplot as plt
plt.show()  
                         
print(graph)
```

```python
def artists_by_popularity(df, popularity_threshold):
    """Devuelve una lista de artistas cuyas canciones tienen una popularidad promedio por encima del umbral dado."""
    popular_artists = df.groupby('artists')['popularity'].mean()
    filtered_artists = popular_artists[popular_artists > popularity_threshold]
    return filtered_artists.index.tolist()
print(artists_by_popularity(df, 80))  # Ejemplo de uso con umbral de popularidad 80
```

```python
def track_name_by_taylor_swift(df):
    """Devuelve una lista de títulos de canciones del artista 'Taylor Swift'."""
    filtered_df = df[df['artists'].str.contains('Taylor Swift', case=False, na=False)]
    return filtered_df['track_name'].tolist()
print(track_name_by_taylor_swift(df))
```

```python
#crea funcion que devuelva la cancion mas popular de un artista dado
def most_popular_song_by_artist(df, artist_name):
    """Devuelve la canción más popular de un artista dado."""
    filtered_df = df[df['artists'].str.contains(artist_name, case=False, na=False)]
    if not filtered_df.empty:
        most_popular_song = filtered_df.loc[filtered_df['popularity'].idxmax()]
        return most_popular_song
    else:
        return None
    
print(most_popular_song_by_artist(df, 'Taylor swift'))  # Ejemplo de uso con el artista 'taylor swift'
    
```
